```python
total_quantity = full_dataset.groupby("PROD_NAME").sum()["PROD_QTY"].sort_values(ascending = False)
brand = {}
for i, t in total_quantity.items():
    new_word = str(i.split()[0])
    if new_word not in brand.keys():
        brand[new_word] = t
    else:
        brand[new_word] += t
        
from collections import OrderedDict
from operator import itemgetter
sort_brand = OrderedDict(sorted(brand.items(), key = itemgetter(1)))
sort_brand
```

```python
pack_size = {}
for i, t in total_quantity.items():
    new_word = str(i.split()[-1][-4:])
    if new_word not in pack_size.keys():
        pack_size[new_word] = t
    else:
        pack_size[new_word] += t

df = pd.DataFrame(list(pack_size.items()),columns = ['pack_size','total_amount_purchased']) 
df
```

```python
df["pack_size"] = df["pack_size"].str.replace("G", "g")
df["pack_size"] = df["pack_size"].str.replace("Salt", "135g")
```

```python
df
```

```python
packet_table =df.groupby("pack_size")["total_amount_purchased"].agg(sum)
```

```python
packet_table_dataframe = pd.DataFrame(data = packet_table).sort_values(by = "total_amount_purchased").reset_index()
```

```python
plt.figure(figsize = (20,20))
plt.barh(packet_table_dataframe["pack_size"], packet_table_dataframe["total_amount_purchased"])
plt.title("Total packet_size purchase by customer")
plt.show()
```

we can see that 175g is the most common pack size purchased from customers, with 150g coming in second, note salt is 135g, as the pack size was not at the end of the PROD_NAME
